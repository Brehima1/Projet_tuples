```python
#1

lila = ("Lila" , "mage" , 100)
torn = ("Torn" , "guerrier" , 150)
eryn = ("Eryn" , "archer"  , 120)

personnages = [lila , torn , eryn ]
print(personnages )
```

```python
#2
personnages[1] = ("Torn" , "guerrier" , 130)
print(personnages)

torn_maj =(torn[0 ], torn[1] , torn[2]-20)
personnages[1] = torn_maj
print(personnages)
```
